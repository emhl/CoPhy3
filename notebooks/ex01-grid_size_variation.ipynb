# Simulation of the Ising modell for zero field an different grid sizes

## setup

```julia
include("../src/ising.jl")
import Plots
```

## temperature sweeps
all simulations ran with the default parameters:
- 100_000 Thermalization steps
- 1_000 Sampling steps

### 1x1x1 grid

```julia
grid_size = 1
J= 1.0
energies, magnetisations, temps = temp_sweep(grid_size=grid_size,J=J)
```

```julia
p1 = Plots.plot(temps, energies[1], color=:blue, label="Energy", seriestype=:scatter, xlabel="temperature (K)", ylabel="energy");
p2 = Plots.plot(temps, magnetisations[1], color=:red, label="Magnetisation", seriestype=:scatter, xlabel="temperature (K)", ylabel="magnetisation");

Plots.plot(p1, p2, layout=(2,1), legend=false)
```
