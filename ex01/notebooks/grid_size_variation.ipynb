# Testing the model without a magnetic field

## setup

```julia
include("../src/ising.jl")
using CairoMakie
```

## temperature sweeps
all ran with the default parameters:
- 100_000 Thermalization steps
- 1_000 Sampling steps

### 1x1x1 grid

```julia
grid_size = 1
J= 1.0
energies, magnetisations, temps = temp_sweep(grid_size=grid_size,J=J)

fig = Figure()
scatter(fig[1, 1], temps, energies[1], color=:blue, label="Energy")
scatter(fig[2, 1], temps, magnetisations[1], color=:red, label="Magnetisation")
fig
```
