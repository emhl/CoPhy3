# Testing the model without a magnetic field

## setup

```julia
include("src/ising.jl")
using CairoMakie
```

```julia
J = 1.0 # coupling constant
B = 0. # magnetic field
N = 100_000 # number of Monte Carlo steps
```

## Temperature between 0K and 10K

```julia
temps , energies, magnetisations = Float64[], Float64[], Float64[]
for T in range(0,15,250)
    grid = create_grid(8,8,8)
    for i in 1:N
        grid = metropolis_step(grid,J,T,B)
    end
    push!(energies, energy(grid,J,B))
    push!(magnetisations, magnetisation(grid))
    push!(temps, T)
end
```

```julia
scatter(temps, energies, color=:blue, label="Energy")
```

```julia
scatter(temps, magnetisations, color=:red, label="Magnetisation")
```
